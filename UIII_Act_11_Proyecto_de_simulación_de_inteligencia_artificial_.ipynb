#### Psicólogo: Enfoque terapéutico, número de pacientes, consultorio, duración de sesión, acreditación y casos cerrados.

```python
# 1. Definición de la lista principal (Base de datos)
biblioteca_psicologos = []

# 2. Función para mostrar la tabla de datos
def mostrar_tabla():
    if not biblioteca_psicologos:
        print("\n[!] La base de datos está vacía actualmente.")
    else:
        print("\n" + "="*115)
        # Encabezados con formato de espacios para alineación
        print(f"{'Enfoque':<20} | {'Pacientes':<10} | {'Consultorio':<15} | {'Duración':<10} | {'Acreditación':<15} | {'Casos Cerrados':<10}")
        print("-" * 115)

        for p in biblioteca_psicologos:
            print(f"{p['enfoque']:<20} | {p['pacientes']:<10} | {p['consultorio']:<15} | {p['duracion']:<10} | {p['acredita']:<15} | {p['cerrados']:<10}")
        print("="*115 + "\n")

# 3. Función para "Simular Inteligencia" (Recomendador)
def asistente_ia():
    print("\n--- Modo Asistente IA Activo ---")
    if not biblioteca_psicologos:
        print("No tengo datos para analizar.")
        return

    especialidad = input("¿Qué tipo de terapia buscas? (Ej: Cognitivo, Gestalt): ")
    encontrado = False

    for p in biblioteca_psicologos:
        if especialidad.lower() in p['enfoque'].lower():
            print(f"-> IA Sugiere: El psicólogo en el consultorio {p['consultorio']} tiene {p['cerrados']} casos exitosos.")
            encontrado = True

    if not encontrado:
        print("Lo siento, no encontré un perfil que coincida con esa búsqueda.")

# 4. Ciclo Principal (Menú Interactivo)
while True:
    print("--- SISTEMA DE GESTIÓN DE SALUD MENTAL ---")
    print("1. Registrar nuevo Psicólogo")
    print("2. Ver Directorio (Tabla)")
    print("3. Consultar a la IA")
    print("4. Salir")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        # Entrada de datos (Key: Value)
        enfoque = input("Enfoque terapéutico: ")
        pacientes = input("Número de pacientes actuales: ")
        consultorio = input("Número o nombre del consultorio: ")
        duracion = input("Duración de sesión (min): ")
        acreditacion = input("Acreditación (Cédula/Título): ")
        cerrados = input("Casos cerrados (éxito): ")

        # Creación del diccionario
        psicologo = {
            "enfoque": enfoque,
            "pacientes": pacientes,
            "consultorio": consultorio,
            "duracion": duracion,
            "acredita": acreditacion,
            "cerrados": cerrados
        }

        # Guardar en la lista
        biblioteca_psicologos.append(psicologo)
        print("\n[OK] Registro guardado con éxito.\n")

    elif opcion == "2":
        mostrar_tabla()

    elif opcion == "3":
        asistente_ia()

    elif opcion == "4":
        print("Cerrando sistema... ¡Hasta luego!")
        break

    else:
        print("\n[!] Opción no válida, intenta de nuevo.\n")
```
