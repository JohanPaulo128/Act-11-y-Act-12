### Numero de lista = 04
#### Psicólogo: Enfoque terapéutico, número de pacientes, consultorio, duración de sesión, acreditación y casos cerrados.

```python
class Psicologo:
    # Esta lista vive dentro de la clase y guardará a todos los psicólogos creados
    lista_psicologos = []

    def __init__(self, enfoque, pacientes, consultorio, duracion, acreditacion, cerrados):
        # Atributos del objeto
        self.enfoque = enfoque
        self.pacientes = pacientes
        self.consultorio = consultorio
        self.duracion = duracion
        self.acreditacion = acreditacion
        self.cerrados = cerrados

    # --- FUNCIONES DENTRO DE LA CLASE ---


    def agregar_medico(cls):
        print("\n--- Registro de Nuevo Psicólogo ---")
        enf = input("Enfoque terapéutico: ")
        pac = input("Número de pacientes: ")
        con = input("Consultorio: ")
        dur = input("Duración de sesión: ")
        acr = input("Acreditación: ")
        cer = input("Casos cerrados: ")

        # Creamos el objeto y lo guardamos en la lista de la clase
        nuevo_p = cls(enf, pac, con, dur, acr, cer)
        cls.lista_psicologos.append(nuevo_p)
        print("✅ Registrado con éxito.")


    def listar_medicos(cls):
        if not cls.lista_psicologos:
            print("\n⚠️ No hay psicólogos registrados.")
            return

        print("\n" + "="*115)
        print(f"{'ID':<4} | {'ENFOQUE':<20} | {'PACIENTES':<10} | {'CONSULTORIO':<20} | {'DURACIÓN':<12} | {'CASOS CERRADOS':<15}")
        print("-" * 115)

        indice = 0
        for ps in cls.lista_psicologos:
            print(f"{indice:<4} | {ps.enfoque:<20} | {ps.pacientes:<10} | {ps.consultorio:<20} | {ps.duracion:<12} | {ps.cerrados:<15}")
            indice += 1
        print("="*115)


    def actualizar_medico(cls):
        cls.listar_medicos()
        if cls.lista_psicologos:
            idx = int(input("\nEscribe el ID del psicólogo a modificar: "))
            print("--- Ingresa los nuevos datos ---")
            cls.lista_psicologos[idx].enfoque = input("Nuevo Enfoque: ")
            cls.lista_psicologos[idx].pacientes = input("Nuevos Pacientes: ")
            cls.lista_psicologos[idx].consultorio = input("Nuevo Consultorio: ")
            cls.lista_psicologos[idx].duracion = input("Nueva Duración: ")
            cls.lista_psicologos[idx].acreditacion = input("Nueva Acreditación: ")
            cls.lista_psicologos[idx].cerrados = input("Nuevos Casos Cerrados: ")
            print("🔄 Datos actualizados.")


    def borrar_medico(cls):
        cls.listar_medicos()
        if cls.lista_psicologos:
            idx = int(input("\nEscribe el ID del psicólogo a eliminar: "))
            cls.lista_psicologos.pop(idx)
            print("🗑️ Registro eliminado.")

# --- MENÚ INTERACTIVO ---

while True:
    print("\n🧠 SISTEMA DE GESTIÓN PSICOLÓGICA 🧠")
    print("1. ➕ Agregar Psicólogo")
    print("2. 📋 Listar Psicólogos")
    print("3. 📝 Actualizar Datos")
    print("4. ❌ Borrar Registro")
    print("5. 🚪 Salir")

    opcion = input("\nSelecciona una opción: ")

    if opcion == "1":
        Psicologo.agregar_medico()
    elif opcion == "2":
        Psicologo.listar_medicos()
    elif opcion == "3":
        Psicologo.actualizar_medico()
    elif opcion == "4":
        Psicologo.borrar_medico()
    elif opcion == "5":
        print("Saliendo del programa... ¡Adiós!")
        break
    else:
        print("Opción no válida, intenta de nuevo.")
```
