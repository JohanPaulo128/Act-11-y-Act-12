#### Psicólogo: Enfoque terapéutico, número de pacientes, consultorio, duración de sesión, acreditación y casos cerrados.

```python
class Psicologo:
    def __init__(self, enfoque, pacientes, consultorio, duracion, acreditacion, cerrados):
        self.enfoque = enfoque
        self.pacientes = pacientes
        self.consultorio = consultorio
        self.duracion = duracion
        self.acreditacion = acreditacion
        self.cerrados = cerrados

psicologos = []

def listar_medicos():
    print(f"\n{'ID':<3} | {'ENFOQUE':<15} | {'PACIENTES':<10} | {'CONSULTORIO':<15}")
    print("-" * 50)
    for i, ps in enumerate(psicologos):
        print(f"{i:<3} | {ps.enfoque:<15} | {ps.pacientes:<10} | {ps.consultorio:<15}")

def agregar_medico():
    psicologos.append(Psicologo(input("Enfoque: "), input("Pacientes: "), input("Consultorio: "), input("Duración: "), input("Acreditación: "), input("Casos cerrados: ")))

def actualizar_medico():
    listar_medicos()
    idx = int(input("ID a actualizar: "))
    psicologos[idx] = Psicologo(input("Nuevo Enfoque: "), input("Nuevos Pacientes: "), input("Nuevo Consultorio: "), input("Nueva Duración: "), input("Nueva Acreditación: "), input("Nuevos Casos: "))

def borrar_medico():
    listar_medicos()
    idx = int(input("ID a borrar: "))
    psicologos.pop(idx)

while True:
    print("\n1. Agregar | 2. Listar | 3. Actualizar | 4. Borrar | 5. Salir")
    op = input("Opción: ")
    if op=="1": agregar_medico()
    elif op=="2": listar_medicos()
    elif op=="3": actualizar_medico()
    elif op=="4": borrar_medico()
    elif op=="5": break
```
